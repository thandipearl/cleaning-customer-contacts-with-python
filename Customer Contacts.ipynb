# Removing Duplicates

```python
df = df.drop_duplicates()
df
```

# Cleaning the Last_Name column

```python
df["Last_Name"] = df["Last_Name"].str.replace(r"[^a-zA-Z]", "", regex=True)
```

```python
df
```

# Removing unnessary columns

```python
df = df.drop(columns = ["Not_Useful_Column"])
df
```

```python
df
```

# Cleaning the Phone_Number Column

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r"\D", "", regex=True)
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(
    lambda x: f"{x[:3]}-{x[3:6]}-{x[6:]}" if pd.notnull(x) and len(x) == 10 else " "
)
```

```python
df
```

# Organizing the address column

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(",", expand=True)
```

```python
df = df.drop(columns=["Address"])
```

```python
df
```

# Standardize Yes/No Columns

```python
df["Paying Customer"] = df["Paying Customer"].replace({
    "Yes": "Y", "No": "N", "Y": "Y", "N": "N"
})
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({
    "Yes": "Y", "No": "N", "Y": "Y", "N": "N"
})
```

```python
df
```

# Handle Missing Values

```python
df = df.fillna("")
```

```python
df = df.replace("N/a", "")
```
